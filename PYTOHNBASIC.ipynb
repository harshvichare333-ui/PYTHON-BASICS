2. What is an interpreter in Python
  -> A Python interpreter is a program that reads and executes Python code line by line. It converts the Python instructions written by a programmer into machine-understandable actions so that the computer can run the program.
  Unlike compiled languages, Python does not convert the entire program into machine code before running. The interpreter executes the code during runtime.

  5. What is mutability in Python
  -> Mutability in Python is the ability of an object to be changed or modified after it is created.
  In Python, objects are divided into two types: mutable objects and immutable objects.
  Mutable objects are those whose values or contents can be changed after creation. Examples include lists, dictionaries, and sets.Immutable objects are those whose values cannot be changed after creation. Examples include strings, integers, and tuples.
  Mutability affects how data is stored and modified in memory. Understanding mutability helps programmers manage data changes and avoid unexpected results in programs.

  6. Why are lists mutable, but tuples are immutable
  -> **Lists are mutable because they are designed to allow changes after creation.** The elements of a list can be added, removed, or modified, which makes lists suitable for storing data that may change during program execution.
  Tuples are immutable because their elements cannot be changed after creation.** Tuples are used when data should remain constant and protected from accidental modifications.
  In theory, **lists provide flexibility for changing data, while tuples provide security and consistency by keeping data unchanged.

7. What is the difference between “==” and “is” operators in Python   
-> **The `==` and `is` operators in Python are used for comparison, but they work differently.**

- **`==` operator** is called the **equality operator**. It is used to check whether the values or contents of two objects are the same.

- **`is` operator** is called the **identity operator**. It is used to check whether two variables refer to the same object stored in memory.

**In theory:**  
The `==` operator compares the **value of objects**, whereas the `is` operator compares the **identity of objects** (whether they are the same object in memory).

8. What are logical operators in Python
-> **Logical operators in Python** are used to combine multiple conditions or expressions and return a result based on whether those conditions are **True or False**.

Python provides three logical operators:

1. **`and` operator**

   * Returns True only when **both conditions are True**.

2. **`or` operator**

   * Returns True when **at least one condition is True**.

3. **`not` operator**

   * Reverses the result of a condition. It changes True to False and False to True.

**In theory:**
Logical operators are used in decision-making statements to perform logical operations and control the flow of a Python program based on multiple conditions.


9. What is type casting in Python
-> **Type casting in Python** is the process of **converting one data type into another data type**. It is used when we need to change the format of a value to perform different operations.

Python provides built-in functions for type casting, such as:

* **int()** – converts a value into an integer
* **float()** – converts a value into a floating-point number
* **str()** – converts a value into a string
* **list()** – converts a value into a list

**In theory:**
Type casting allows programmers to convert data from one type to another so that different types of operations can be performed correctly and efficiently in a Python program.

10. What is the difference between implicit and explicit type casting
- > **The difference between implicit and explicit type casting is based on how the conversion of data types is performed.**

* **Implicit Type Casting:**
  Implicit type casting is the automatic conversion of one data type into another by Python without any instruction from the programmer. It usually happens when Python converts a smaller or compatible data type into a larger data type to avoid data loss.

* **Explicit Type Casting:**
  Explicit type casting is the manual conversion of one data type into another by the programmer using built-in functions such as **int(), float(), str(),** etc.

**In theory:**
Implicit type casting is performed automatically by Python, whereas explicit type casting is performed manually by the programmer when a specific data type conversion is required.

11. What is the purpose of conditional statements in Python
-> **Conditional statements in Python** are used to make decisions in a program based on certain conditions. They allow the program to execute different blocks of code depending on whether a condition is **True or False**.

Python uses conditional statements such as **if, elif, and else** to control the flow of execution.

**In theory:**
The purpose of conditional statements is to provide decision-making ability in programs by checking conditions and executing appropriate actions based on the results. They help create flexible and logical programs.

12. How does the elif statement work
-> **The `elif` statement in Python** is used to check multiple conditions after an `if` statement. It stands for **"else if"** and allows a program to test another condition when the previous condition is False.

The `elif` statement is executed only when the conditions before it are not satisfied. A program can have multiple `elif` statements, but only the first True condition block will be executed.

**In theory:**
The `elif` statement provides additional decision-making options in a conditional structure by allowing multiple conditions to be checked in sequence. It helps control the flow of a program when there are multiple possible outcomes.


13. What is the difference between for and while loops
-> **The difference between `for` and `while` loops in Python is based on how they control repetition.**

* **`for` loop:**
  A `for` loop is used to iterate over a sequence of elements such as a list, tuple, string, or range. It is generally used when the number of repetitions is known or when we need to process each item in a collection.

* **`while` loop:**
  A `while` loop is used to execute a block of code repeatedly as long as a specified condition remains True. It is generally used when the number of repetitions is not fixed and depends on a condition.

**In theory:**
A `for` loop is used for iterating over a sequence or a known number of times, whereas a `while` loop is used for repeating tasks based on a condition until the condition becomes False.

14. Describe a scenario where a while loop is more suitable than a for loop.
-> **A `while` loop is more suitable than a `for` loop when the number of repetitions is not known in advance and depends on a condition.**

For example, in a **user login system**, a program may continue asking the user to enter a password until the correct password is provided. Since we do not know how many attempts the user will need, a `while` loop is more appropriate.

**In theory:**
A `while` loop is used in situations where repetition should continue as long as a specific condition is true, while a `for` loop is better when the number of iterations is already known.

```python
#6.
# Function to demonstrate mutable and immutable arguments

def modify_values(number, my_list):

    # Immutable argument
    number = number + 10

    # Mutable argument
    my_list.append(4)

    print("Inside function:")
    print("Number:", number)
    print("List:", my_list)


# Creating values
num = 5
list1 = [1, 2, 3]


# Calling function
modify_values(num, list1)


print("Outside function:")
print("Number:", num)
print("List:", list1)
```

```python
#13.
# Initialize sum
total = 0

# Loop through numbers from 1 to 50
for i in range(1, 51):

    # Check if number is even
    if i % 2 == 0:
        total = total + i

# Display result
print("Sum of even numbers between 1 and 50 is:", total)
```

```python
#14.
# Taking input from user
string = input("Enter a string: ")

# Initialize variables
reverse = ""
index = len(string) - 1

# Using while loop to reverse string
while index >= 0:
    reverse = reverse + string[index]
    index = index - 1

# Display result
print("Reversed string:", reverse)
```

```python
#15.
# Taking input from user
num = int(input("Enter a number: "))

# Initialize factorial
factorial = 1

# Using while loop
while num > 0:
    factorial = factorial * num
    num = num - 1

# Display result
print("Factorial is:", factorial)

```
